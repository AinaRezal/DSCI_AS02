```python
import numpy as np
import pandas as pd

df_eq = pd.read_csv('earthquake.csv')
df_dis = pd.read_csv('natural_disasters.csv')
```

## EARTHQUAKE

```python
df_eq.head()
```

```python
df_eq.dtypes
```

```python
df_eq.isnull().sum()
```

```python
df_eq.drop(['title', 'country', 'continent','location'], axis='columns', inplace=True)

df_eq
```

```python
df_eq['alert'] = df_eq['alert'].fillna('unknown')

df_eq
```

```python
df_eq['date_time'] = pd.to_datetime(df_eq['date_time'], format='%d-%m-%Y %H:%M')
df_eq['year'] = pd.to_datetime(df_eq['date_time']).dt.year
df_eq['month'] = pd.to_datetime(df_eq['date_time']).dt.month
    
df_eq
```

```python
df_eq.duplicated().sum()
```

```python
df_eq.to_csv('cleaned_earthquake.csv')
```

## Natural Disaster

```python
df_dis.head()
```

```python
df_dis.isnull().sum()
```

```python
df_dis.dropna(subset=['Reconstruction costs from earthquakes',
         'Insured damages against earthquakes',
         'Total economic damages from earthquakes',
         'Total economic damages from earthquakes as a share of GDP'],
          inplace=True)
df_dis = df_dis.reset_index()
df_dis = df_dis.drop(['index'], axis=1)
df_dis
```

```python
df_dis = df_dis.rename(columns={
    'Number of deaths from earthquakes' : 'Total deaths',
    'Number of people injured from earthquakes' : 'Total injured',
    'Number of people affected by earthquakes' : 'Total affected',
    'Number of people left homeless from earthquakes' : 'Total homeless',
    'Number of total people affected by earthquakes' : 'Total people',
    'Reconstruction costs from earthquakes' : 'Reconstruction costs',
    'Insured damages against earthquakes' : 'Insured damages',
    'Total economic damages from earthquakes' : 'Total economic damages',
    'Death rates from earthquakes' : 'Death rates',
    'Injury rates from earthquakes' : 'Injury rates',
    'Number of people affected by earthquakes per 100,000' : 'Number of affected people per 100,000',
    'Homelessness rate from earthquakes' : ' Homelessness rate',
    'Total number of people affected by earthquakes per 100,000' : 'Total affected people per 100,000',
    'Total economic damages from earthquakes as a share of GDP' : 'Total economic damages as a share of GDP'
})

df_dis
```

```python
df_dis.to_csv('cleaned_disaster.csv')
```
